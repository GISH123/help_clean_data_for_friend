```python
family_a1_set = set()
family_index_start = 0
family_index_end = 0
family_number = 0
subdata['family_type'] = ''
subdata['family_number'] = 0
for index, row in subdata.iterrows():
    if index ==0:
        family_a1_set.add(row['a1'])
        family_index_start = 0
        continue
    if (row['id2'] == subdata.iloc[index-1]['id2'] +1):
        family_a1_set.add(row['a1'])
    else:
        #如果id2不連續
        family_index_end = index-1
        if(family_a1_set.issubset({'戶長'})):
            for i in range(family_index_start,family_index_end+1):
                subdata.at[i,'family_type'] = '單人家庭' 
        elif(family_a1_set.issubset({'戶長','父母'})):
            for i in range(family_index_start,family_index_end+1):
                subdata.at[i,'family_type'] = '單身有父母家庭'  
        elif(family_a1_set.issubset({'戶長','子女'})):
            for i in range(family_index_start,family_index_end+1):
                subdata.at[i,'family_type'] = '單親家庭'
        elif(family_a1_set.issubset({'戶長','配偶'})):
            for i in range(family_index_start,family_index_end+1):
                subdata.at[i,'family_type'] = '夫婦家庭'
        elif(family_a1_set.issubset({'戶長','配偶','子女'})):
            for i in range(family_index_start,family_index_end+1):
                subdata.at[i,'family_type'] = '核心家庭'
        elif(family_a1_set.issubset({'戶長','配偶','子女','父母'})):
            for i in range(family_index_start,family_index_end+1):
                subdata.at[i,'family_type'] = '折衷家庭'
        elif({'戶長','配偶','父母','子女','兄弟姊妹'}.issubset(family_a1_set)):
            for i in range(family_index_start,family_index_end+1):
                subdata.at[i,'family_type'] = '大家庭'                  
        else:
            for i in range(family_index_start,family_index_end+1):
                subdata.at[i,'family_type'] = '其它'             

                
        for i in range(family_index_start,family_index_end+1):
            subdata.at[i,'family_number'] = family_number   
                
                
        family_index_start = index
        family_a1_set = set()
        family_a1_set.add(row['a1'])
        family_number += 1
#         for i in range(family_index_start,family_index_end+1):
#             subdata.at[i,'family_number'] = family_number       
 

    if(index == len(subdata)-1):
        family_index_end = index
        if(family_a1_set.issubset({'戶長'})):
            for i in range(family_index_start,family_index_end+1):
                subdata.at[i,'family_type'] = '單人家庭' 
        elif(family_a1_set.issubset({'戶長','父母'})):
            for i in range(family_index_start,family_index_end+1):
                subdata.at[i,'family_type'] = '單身有父母家庭'  
        elif(family_a1_set.issubset({'戶長','子女'})):
            for i in range(family_index_start,family_index_end+1):
                subdata.at[i,'family_type'] = '單親家庭'
        elif(family_a1_set.issubset({'戶長','配偶'})):
            for i in range(family_index_start,family_index_end+1):
                subdata.at[i,'family_type'] = '夫婦家庭'
        elif(family_a1_set.issubset({'戶長','配偶','子女'})):
            for i in range(family_index_start,family_index_end+1):
                subdata.at[i,'family_type'] = '核心家庭'
        elif(family_a1_set.issubset({'戶長','配偶','子女','父母'})):
            for i in range(family_index_start,family_index_end+1):
                subdata.at[i,'family_type'] = '折衷家庭'
        elif({'戶長','配偶','父母','子女','兄弟姊妹'}.issubset(family_a1_set)):
            for i in range(family_index_start,family_index_end+1):
                subdata.at[i,'family_type'] = '大家庭'                  
        else:
            for i in range(family_index_start,family_index_end+1):
                subdata.at[i,'family_type'] = '其它'             

        for i in range(family_index_start,family_index_end+1):
            subdata.at[i,'family_number'] = family_number   
            
    if(index%1000==0):
        print(index)
        
```

```python
def head_of_household(family_df):
    family_df['ret'] = family_df['a7']
    family_df['age'] = family_df['a3']
    family_df['work'] = family_df['a23_a']
    family_df['edu'] = family_df['a5_2']
    family_df['wage'] = family_df['b1_a']
    family_df['hr_full'] = family_df['a10_1a'] + family_df['a10_2a']
    family_df['hr_part'] = family_df['a10_2b'] + family_df['a10_2b']
    family_df['ind2'] = family_df['a21_a2']
    family_df['P2'] = family_df['b6_b']
    family_df['add_hr'] = family_df['a12']
    family_df['exp'] = family_df['b3_y']
    family_df['Health1'] = family_df['a8']
    family_df['Health2'] = family_df['a11']
    return family_df
```

```python
def parents(family_df):
    family_df.loc[family_df['b1_a']=='跳答或不適用','b1_a'] = '0'
    family_df.loc[family_df['b1_a']=='虧損','b1_a'] = '0'
    family_df['b1_a'] = family_df['b1_a'].astype('int32')
    family_df['Phealth'] = family_df['a11']
#     family_df['Cwage'] = family_df['b1_a']
    family_df['Pret'] = family_df['a7']
    family_df['Pwage'] = family_df.groupby(['family_number'])['b1_a'].transform('sum')
    family_df['Phr_part'] = family_df['a10_1b']+family_df['a10_2b']
    return family_df
```

```python
def head_of_household_for_phr_full(family_df):
    family_df['ret'] = family_df['a7']
    family_df['age'] = family_df['a3']
    family_df['work'] = family_df['a23_a']
    family_df['edu'] = family_df['a5_2']
    family_df['wage'] = family_df['b1_a']
    family_df['hr_full'] = family_df['a10_1a'] + family_df['a10_2a']
    family_df['hr_part'] = family_df['a10_1b'] + family_df['a10_2b']
    family_df['ind2'] = family_df['a21_a2']
    family_df['P2'] = family_df['b6_b']
    family_df['add_hr'] = family_df['a12']
    family_df['exp'] = family_df['b3_y']
    family_df['Health1'] = family_df['a8']
    family_df['Health2'] = family_df['a11']
    family_df['Phr_full'] = family_df[family_df['a1']=='父母']['a10_1a'] + family_df[family_df['a1']=='父母']['a10_2a']
    return family_df
```

```python
def head_of_household_child(family_df):
    family_df['Child_1'] = family_df['b17_a1']
    family_df['Child_2'] = family_df['b17_a2']
    family_df['Child_3'] = family_df['b17_a3']
    family_df['Child_4'] = family_df['b17_a4']
    family_df['Child_5'] = family_df['b17_a5']
    family_df['Child_6'] = family_df['b17_a6']
#     family_df['Child_7'] = family_df['b17_a7'] 資料沒這欄位
    return family_df
```

```python
def child(family_df):
    family_df['a5_1'] = family_df['a5_1'].str.split("(",1).str[0]
    family_df['Ungraduated_child'] = family_df.groupby(['family_number','a5_1'])['a5_1'].transform('count')
    family_df['Cage'] = family_df['a3']
    a5_2_df = family_df.groupby('family_number')['a5_2'].apply(','.join).reset_index()
    a5_2_df['Cedu'] = a5_2_df['a5_2']
    del a5_2_df['a5_2']
    
    temp_df = family_df[['family_number','a5_2']]
    temp_df['Cedu'] = temp_df['a5_2']
    del temp_df['a5_2']
    
    family_df = family_df.reset_index(drop=True)
    family_df['Cedu'] = temp_df.merge(a5_2_df,on='family_number')['Cedu_y']
    
    family_df['Chealth'] = family_df['a11']
    family_df['Chealth1'] = family_df['a8']
    family_df['Chealth2'] = family_df['a11']
    family_df['Chr_full'] = family_df['a10_1a'] + family_df['a10_2a']
    family_df['Chr_part'] = family_df['a10_1b'] + family_df['a10_2b']
    family_df['Cwage'] = family_df.groupby(['family_number'])['b1_a'].transform('sum')
    
    
    
    
    
    return family_df
```

```python
one_man_with_child_family_b_processed = child(one_man_with_child_family_b)
```
